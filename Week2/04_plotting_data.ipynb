```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 0.98, 0.01)
```

```python
y1 = np.sin(2 * np.pi * 4 * t)
```

```python
sns.scatterplot(x=t, y=y1)
```

```python
y2 = np.cos(2 * np.pi * 4 * t)
```

```python
sns.scatterplot(x=t, y=y2)
```

```python
fig, ax = plt.subplots()
ax.set(title='my plot', xlabel='time', ylabel='value')
sns.scatterplot(x=t, y=y1, ax=ax, label='sin')
sns.scatterplot(x=t, y=y2, ax=ax, label='cos')
```

```python
fig.savefig('myPlot.png')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=False, sharey=False)
sns.scatterplot(x=t, y=y1, ax=ax1)
sns.scatterplot(x=t, y=y2, ax=ax2)
```
